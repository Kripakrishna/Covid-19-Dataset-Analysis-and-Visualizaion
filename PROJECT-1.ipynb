```python
#total active cases

data1["Total_Active_cases"]=data1["Confirmed"]-(data1["Cured"]+data1["Deaths"])
```

```python
statewise=pd.pivot_table(data1,values=["Confirmed","Deaths","Cured"],index="State/UnionTerritory",aggfunc=max)
```

```python
statewise["Recovery_Rate"]=(statewise["Cured"]/statewise["Confirmed"])*100
```

```python
#remove rows where state=india

vaccine=data2[data2.State!='India']
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
vaccine.head()
```

```python
#Most vaccinated state

max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python

fig=plt.figure(figsize=(16,9))
plt.title("Top 5 vaccinated states in india", size=25)
ax=sns.barplot(data=max_vac.iloc[:5],y=max_vac.Total,x=max_vac.index,linewidth=2, edgecolor='red')
#plt.xlabel("States")
#plt.ylabel("Vaccination")

plt.show()
```

```python
#minimum vaccinated states
min_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac=min_vac.sort_values('Total',ascending=True)[:5]
min_vac
```

```python

fig=plt.figure(figsize=(16,9))
plt.title("Top 5 non vaccinated states in india", size=25)
ax=sns.barplot(data=min_vac.iloc[:10],y=min_vac.Total,x=min_vac.index,linewidth=2, edgecolor='red')
#plt.xlabel("States")
#plt.ylabel("Vaccination")

plt.show()
```
